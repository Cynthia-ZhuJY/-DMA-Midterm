## matplotlib - a visualization lib

# Feb. 22 Pandas

### creating a series

```python
import pandas as pd
pd.Series([3, 4, 5])
```

### adding labels

```python
import pandas as pd
pd.Series([3, 4, 5], list('abc'))
```

```python
pd.Series([3, 4, 5], index = list('abc'))
```

```python
pd.Series({'a':1, 'b':2, 'c':3})
```

### calculating specific indexes

```python
pd.Series({'a':1, 'b':2, 'c':3}, index = ['a', 'c'])
```

### null values

```python
s = pd.Series({'x':100}, ['x', 'y'])
s.isnull()
```

```python
s.notnull()
```

### indexing in pandas

#### indexing with duplicate lables

```python
s = pd.Series([7, 8, 9, 10], list('xyxz'))
s['x'] 
```

#### vectorized variables

```python
s = pd.Series([1,2], ['x', 'y'])
s + 1
```

#### label alignment

```python
s = pd.Series([1,2], ['x', 'y'])
s + pd.Series([9,100], ['x', 'y'])
```

```python
# tricky!
s = pd.Series([1,2], ['x', 'y'])
s + pd.Series([9,100], ['x', 'z'])
```

### filtering with pandas

```python
s = pd.Series([2, 3, 4, 5])
s % 2
```

```python
s % 2 == 1
```

#### filtering out odd numbers using indexing

```python
s[s % 2 == 1]
```
