character classes:

* `[^a] - not an a
* `[a|b]` - a or b
* `[0-9]` - 0 through 9
* `[a-z]` - lowercase a - z
* `.` - any character (except newlines)
* `\d` - digit [0-9]
* `\s` - space character
* `\w` - word character  [0-9A-Za-z_]
